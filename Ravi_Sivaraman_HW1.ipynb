```python
#Grade 5 books
#Wizard of OZ
wizard_of_oz_url = "https://www.gutenberg.org/files/55/55-0.txt"

wizard_of_oz_stories_booktext = getBookasText(wizard_of_oz_url)

lexical_diversity(wizard_of_oz_stories_booktext)
```

```python
    
def getBookasText(bookurl):
    filedata = urllib2.urlopen(bookurl)
    booktext = ""
    for eachline in filedata: 
        booktext = booktext + eachline.decode("utf-8")
    return booktext
```

```python
vocab_size_animal = len(set(gutenberg_animal_book))
vocab_size_ppj = len(set(pride_and_pj_booktext))
vocab_size_wizard = len(set(wizard_of_oz_stories_booktext))
vocab_size_grade_1_book = len(set(grade_level_1_stories_booktext))


print("Vocabulary size of Animal book", vocab_size_animal)
print("Vocabulary size of Pride and Prejuice book", vocab_size_ppj)
print("Vocabulary size of Wizard of Oz book", vocab_size_wizard)
print("Vocabulary size of Grade 1 Stories book", vocab_size_grade_1_book)
```
